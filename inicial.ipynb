### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://drive.google.com/drive/folders/1uDesZePdkhiraJmiyeZ-w5tfc8XsNYFZ?usp=drive_link

```python
# !pip install pandas numpy openpyxl nbformat ipykernel plotly
import pandas as pd
tabela = pd.read_csv("cancelamentos_sample.csv")
# deletar coluna dentro de uma tabela de consulta
# tabela = tabela.drop(columns=["CustomerID", "idade", "sexo"])
tabela = tabela.drop(columns="CustomerID")
tabela = tabela.dropna() # Remover valores vazio ( comando .dropna)


# print(tabelas) -> imprime na tela em python
display(tabela.info())


display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format)) # Normalize mostra em % # corrige as casas decimais % .map("{:.1%}".format



import plotly.express as px # biblioteca para usar graficos

# criar variavel para exibir varias comparações de colunas
coluna = "ligacoes_callcenter"
#for coluna in tabela.columns:
grafico = px.histogram(tabela, x=coluna, color="cancelou")
    
     # criar o grafico, tabela vai mostrar no x todas idades e dividir por cor cancelamento
grafico.show() # e o print da tela.
```
